```python
# Step 1: Import Libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Step 2: Load Dataset (created earlier)
df = pd.read_csv('Pakistan_Unemployment.csv')
print("ðŸ”¹ First 5 rows:\n", df.head())

# Step 3: Check Info & Summary
print("\nðŸ”¹ Dataset Info:")
print(df.info())

print("\nðŸ”¹ Null Values:\n", df.isnull().sum())

# Step 4: Unemployment Trend Over Time
plt.figure(figsize=(12, 6))
sns.lineplot(data=df, x='Year', y='Unemployment Rate', marker='o', color='darkblue')
plt.title('Unemployment Rate in Pakistan (2000â€“2023)', fontsize=14)
plt.xlabel('Year')
plt.ylabel('Unemployment Rate (%)')
plt.grid(True)
plt.tight_layout()
plt.show()

# Step 5: Correlation Heatmap
plt.figure(figsize=(6, 5))
sns.heatmap(df.corr(), annot=True, cmap='Purples')
plt.title('Correlation Matrix')
plt.tight_layout()
plt.show()
```
